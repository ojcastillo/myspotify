## Boilerplate code

```python
import json

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set();
sns.set_style("whitegrid");
%matplotlib inline
```

## Read downloaded library

Assuming the library has been downloaded already, load the data directly from the json file:

```python
with open('../assets/user_library.json') as json_f:
    json_data = json.load(json_f)
```

```python
print(f'Number of trackes loaded: {len(json_data)}')
```

## Transforming data

Flatten the json so is possible to create a DataFrame from it:

```python
df = pd.json_normalize(json_data)
df.head()
```

```python
print(f'Shape of data frame: {df.shape}')
print(f'Columns: {df.columns}')
```

## Exploring library

### Number of tracks per year

We have a `track.album.release_date` field which should be possible to use:

```python
df['track.album.release_date'].head()
```

We need extract the release year portion and generate counts per year for plotting them:

```python
df['track.album.release_year'] = df['track.album.release_date'].str.split('-').str.get(0).astype(int)
df['track.album.release_year'].head()
```

```python
plt.figure(figsize=(16, 6))
ax = sns.countplot(data=df, x="track.album.release_year")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right");
```

### Number of tracks per decade

```python
df['track.album.release_century'] = (df['track.album.release_year'] // 100) * 100
df['track.album.release_century'].value_counts().sort_index(ascending=False)
```

```python
df['track.album.release_decade_in_century'] = ((df['track.album.release_year'] % 100) // 10) * 10
df['track.album.release_decade'] = df['track.album.release_century'] + df['track.album.release_decade_in_century']
```

```python
plt.figure(figsize=(16, 6))
ax = sns.countplot(data=df, x="track.album.release_decade")
```

### Top 50 artists by saved track counts

```python
df['track.artists'] = df['track.album.artists'].apply(lambda artists: [a['name'] for a in artists])
```

```python
sorted_artists = df.explode('track.artists')['track.artists'].value_counts().sort_values(ascending=False)
top50_artists = sorted_artists.head(50)
```

```python
plt.figure(figsize=(16, 6))
ax = sns.barplot(x=top50_artists.index, y=top50_artists.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right");
```

### Top artists that released music in 2021

```python
def get_top_artist_from_year(data, year, col_name="track.album.release_year", topk=50):
    sorted_artists = (
        data[data[col_name] == year]
            .explode('track.artists')['track.artists']
            .value_counts()
            .sort_values(ascending=False)
    )
    return sorted_artists.head(topk)
```

```python
top50_artists21 = get_top_artist_from_year(df, 2021)
```

```python
plt.figure(figsize=(16, 6))
ax = sns.barplot(x=top50_artists21.index, y=top50_artists21.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right");
```

### Number of tracks per year of addition to library

```python
df["added_at.year"] = df["added_at"].str.split('-').str.get(0).astype(int)
df["added_at.year"].head()
```

```python
plt.figure(figsize=(16, 6))
ax = sns.countplot(data=df, x="added_at.year");
```

### Get 2021 top artists per number of additions to library

```python
top50_added_artists21 = get_top_artist_from_year(df, 2021, col_name="added_at.year")
```

```python
plt.figure(figsize=(16, 6))
ax = sns.barplot(x=top50_added_artists21.index, y=top50_added_artists21.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right");
```

## Enriching library with audio features

```python
with open('../assets/audio_features_orlando_1266569549.json') as json_f:
    json_data = json.load(json_f)
```

```python
print(f'Number of audio features loaded: {len(json_data)}')
```

```python
features_df = pd.json_normalize(json_data)
features_df.head()
```

```python
clean_features_df = (
    features_df
        .rename(columns={col: f"track.{col}" for col in features_df.columns})
        .drop(columns=["track.type", "track.uri", "track.duration_ms"])
)
clean_features_df.columns
```

```python
joined_df = df.merge(clean_features_df, on="track.id")
joined_df.columns
```

```python
print(f'Size of join table: {len(joined_df)}')
```

## Exploring library with audio features

### Various histograms

```python
joined_df["track.duration_sec"] = (joined_df["track.duration_ms"] / 1000.0)
joined_df["track.duration_min"] = (joined_df["track.duration_sec"] / 60.0)
```

```python
fig, axs = plt.subplots(nrows=4, ncols=2)
fig.set_size_inches(16, 24);
sns.histplot(data=joined_df, x="track.duration_min", binwidth=1.0, ax=axs[0][0])
sns.histplot(data=joined_df, x="track.danceability", ax=axs[0][1])
sns.histplot(data=joined_df, x="track.energy", ax=axs[1][0])
sns.histplot(data=joined_df, x="track.loudness", ax=axs[1][1])
sns.histplot(data=joined_df, x="track.acousticness", ax=axs[2][0])
sns.histplot(data=joined_df, x="track.instrumentalness", ax=axs[2][1])
sns.histplot(data=joined_df, x="track.liveness", ax=axs[3][0])
sns.histplot(data=joined_df, x="track.valence", ax=axs[3][1])
```
